# Deep Neural Network

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
np.random.seed(0)
```

```python
n_pts = 500
X, y = datasets.make_circles(n_samples=n_pts, random_state=123, noise=0.1, factor=0.20)
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
plt.show()
```

# This is using 3 hidden layers to train the model.

```python
model = Sequential()
model.add(Dense(3, input_shape=(2,), activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
model.compile(Adam(learning_rate=0.03), 'binary_crossentropy', metrics=['accuracy'])
```

```python
h = model.fit(x=X, y=y, verbose=1, batch_size=20, epochs=100, shuffle='true')
```

```python
plt.plot(h.history['accuracy'])
plt.xlabel('Epoch')
plt.legend(['Accuracy'])
plt.title('Accuaracy with 3 hidden layers')
```

```python
plt.plot(h.history['loss'])
plt.xlabel('Epoch')
plt.legend(['Loss'])
plt.title('Loss with 3 hidden layers')
```

## Plot decision boundery for 3 hidden layers

```python
def plot_decision_boundery(X, y, model):
    x_span = np.linspace(min(X[:, 0])-1, max(X[:, 0])+1, 50)
    y_span = np.linspace(min(X[:, 1])-1, max(X[:, 1])+1, 50)
    xx, yy = np.meshgrid(x_span, y_span)
    xx_, yy_ = xx.ravel(), yy.ravel()
    grid = np.c_[xx_, yy_]
    pred_func =  model.predict(grid)
    z = pred_func.reshape(xx.shape)
    plt.contourf(xx, yy, z)
    plt.scatter(X[:n_pts, 0], X[:n_pts, 1])
    plt.scatter(X[n_pts:, 0], X[n_pts:, 1])
```

```python
plot_decision_boundery(X, y, model)
x = 0
y = 2
point = np.array([[x, y]])
prediction = model.predict(point)
plt.plot([x], [y], marker="o", markersize=10, color="red")
print(str(np.round(prediction*100)) + "%")
```

## Model for 4 hidden layers

```python
model = Sequential()
model.add(Dense(4, input_shape=(2,), activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
model.compile(Adam(learning_rate=0.03), 'binary_crossentropy', metrics=['accuracy'])
```

```python
h = model.fit(x=X, y=y, verbose=1, batch_size=20, epochs=100, shuffle='true')
```

```python
plt.plot(h.history['accuracy'])
plt.xlabel('Epoch')
plt.legend(['Accuracy'])
plt.title('Accuaracy with 4 hidden layers')
```

```python
plt.plot(h.history['loss'])
plt.xlabel('Epoch')
plt.legend(['Loss'])
plt.title('Loss with 4 hidden layers')
```

```python
plot_decision_boundery(X, y, model)
x = 0
y = 0.6
point = np.array([[x, y]])
prediction = model.predict(point)
plt.plot([x], [y], marker="o", markersize=10, color="red")
print(str(np.round(prediction*100)) + "%")
```
